# Calculation of the diffraction angles for DVD, CD and a diffraction grating of 1000 l/mm



## The incident and final angles are measured wrt the normal of the diffraction grating

The incident angle is positive to the left and positive to t:he right of the normal line

The final angle is positive to the right and positive to the left of the normal line

```
import matplotlib.pyplot as plt
import numpy as np
import math
theta_i = 70 # incident angle of light beam wrt normal
d_DVD = 740 # nm DVD
#d_BLU = 320 # nm Blueray
d_1000 = 1000 # nm
d_CD = 1600 # nm DVD
lamda_R = 635 # nm
lamda_G = 550 # nm
lamda_B = 450 # nm

Phase_R_DVD = []
Phase_G_DVD = []
Phase_B_DVD = []
Phase_R_CD = []
Phase_G_CD = []
Phase_B_CD = []
Phase_R_1000 = []
Phase_G_1000 = []
Phase_B_1000 = []
Theta = []
A = []
phase_R_DVD_old = 10
phase_G_DVD_old = 10
phase_B_DVD_old = 10
phase_R_CD_old = 10
phase_G_CD_old = 10
phase_B_CD_old = 10
phase_R_1000_old = 10
phase_G_1000_old = 10
phase_B_1000_old = 10


for theta in range(-90,90):
  phase = np.sin(theta_i*np.pi/180)-np.sin(theta*np.pi/180)
  phase_R_DVD = phase*d_DVD/lamda_R
  phase_G_DVD = phase*d_DVD/lamda_G
  phase_B_DVD = phase*d_DVD/lamda_B
  phase_R_CD = phase*d_CD/lamda_R
  phase_G_CD = phase*d_CD/lamda_G
  phase_B_CD = phase*d_CD/lamda_B
  phase_R_1000 = phase*d_1000/lamda_R
  phase_G_1000 = phase*d_1000/lamda_G
  phase_B_1000 = phase*d_1000/lamda_B
  Phase_R_DVD.append(phase*d_DVD/lamda_R)
  Phase_G_DVD.append(phase*d_DVD/lamda_G)
  Phase_B_DVD.append(phase*d_DVD/lamda_B)
  Phase_R_CD.append(phase*d_CD/lamda_R)
  Phase_G_CD.append(phase*d_CD/lamda_G)
  Phase_B_CD.append(phase*d_CD/lamda_B)
  Phase_R_1000.append(phase*d_1000/lamda_R)
  Phase_G_1000.append(phase*d_1000/lamda_G)
  Phase_B_1000.append(phase*d_1000/lamda_B)
  Theta.append(theta)
  if (np.abs(math.floor(phase_R_DVD) - math.floor(phase_R_DVD_old)) == 1):
    s = "Angle_R_DVD = " + str(theta)
    order = max(math.floor(phase_R_DVD),math.floor(phase_R_DVD_old))
    A.append([s,order])
  if (np.abs(math.floor(phase_G_DVD) - math.floor(phase_G_DVD_old)) ==1 ):
    s = "Angle_G_DVD = " + str(theta)
    order = max(math.floor(phase_G_DVD),math.floor(phase_G_DVD_old))
    A.append([s,order])
  if (np.abs(math.floor(phase_B_DVD) - math.floor(phase_B_DVD_old)) ==1 ):
    s = "Angle_B_DVD = " + str(theta)
    order = max(math.floor(phase_B_DVD),math.floor(phase_B_DVD_old))
    A.append([s,order])
  if (np.abs(math.floor(phase_R_CD) - math.floor(phase_R_CD_old)) ==1 ):
    s = "Angle_R_CD = " + str(theta)
    order = max(math.floor(phase_R_CD),math.floor(phase_R_CD_old))
    A.append([s,order])
  if (np.abs(math.floor(phase_G_CD) - math.floor(phase_G_CD_old)) ==1 ):
    s = "Angle_G_CD = " + str(theta)
    order = max(math.floor(phase_G_CD),math.floor(phase_G_CD_old))
    A.append([s,order])
  if (np.abs(math.floor(phase_B_CD) - math.floor(phase_B_CD_old)) ==1 ):
    s = "Angle_B_CD = " + str(theta)
    order = max(math.floor(phase_B_CD),math.floor(phase_B_CD_old))
    A.append([s,order])
  if (np.abs(math.floor(phase_R_1000) - math.floor(phase_R_1000_old)) ==1 ):
    s = "Angle_R_1000 = " + str(theta)
    order = max(math.floor(phase_R_1000),math.floor(phase_R_1000_old))
    A.append([s,order])
  if (np.abs(math.floor(phase_G_1000) - math.floor(phase_G_1000_old)) ==1 ):
    s = "Angle_G_1000 = " + str(theta)
    order = max(math.floor(phase_G_1000),math.floor(phase_G_1000_old))
    A.append([s,order])
  if (np.abs(math.floor(phase_B_1000) - math.floor(phase_B_1000_old)) ==1 ):
    s = "Angle_B_1000 = " + str(theta)
    order = max(math.floor(phase_B_1000),math.floor(phase_B_1000_old))
    A.append([s,order])
  phase_R_DVD_old = phase_R_DVD
  phase_G_DVD_old = phase_G_DVD
  phase_B_DVD_old = phase_B_DVD
  phase_R_CD_old = phase_R_CD
  phase_G_CD_old = phase_G_CD
  phase_B_CD_old = phase_B_CD

  phase_R_1000_old = phase_R_1000
  phase_G_1000_old = phase_G_1000
  phase_B_1000_old = phase_B_1000

for i in A:
  if (i[1] == -3):
    print ("Order = -3",i[0])
print()
for i in A:
  if (i[1] == -2):
    print ("Order = -2",i[0])
print()
for i in A:
  if (i[1] == -1):
    print ("Order = -1",i[0])
print()
for i in A:
  if (i[1] == 0):
    print ("Order = 0",i[0])
print()
for i in A:
  if (i[1] == 1):
    print ("Order = 1",i[0])
print()
for i in A:
  if (i[1] == 2):
    print ("Order = 2",i[0])
print()
for i in A:
  if (i[1] == 3):
    print ("Order = 3",i[0])
plt.figure(figsize=(12,10))
plt.plot(Theta,Phase_R_DVD,'-r',label='Red DVD')
plt.plot(Theta,Phase_G_DVD,'-g',label='Green DVD')
plt.plot(Theta,Phase_B_DVD,'-b',label='Blue DVD')
plt.plot(Theta,Phase_R_CD,'--r',label='Red CD')
plt.plot(Theta,Phase_G_CD,'--g',label='Green CD')
plt.plot(Theta,Phase_B_CD,'--b',label='Blue CD')
plt.plot(Theta,Phase_R_1000,':r',label='Red 1000')
plt.plot(Theta,Phase_G_1000,':g',label='Green 1000')
plt.plot(Theta,Phase_B_1000,':b',label='Blue 1000')
plt.title('Incident Angle wrt Normal = '+str(theta_i))
plt.xlabel("Final angle (degree)")
plt.ylabel("Order of Difraction")
plt.legend()
plt.grid(which = "major", linewidth = 1)
plt.grid(which = "minor", linewidth = 0.2)
plt.minorticks_on()
```
